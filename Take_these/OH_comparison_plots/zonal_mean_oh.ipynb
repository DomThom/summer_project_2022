```python
import xarray as xr
import matplotlib.dates as mdates
import numpy as np
import matplotlib
from matplotlib import rcParams
import datetime
import pandas as pd
import cftime
import pylab as plt
import netCDF4 as nc
from numpy import ma
import dask as dd
```

```python
# define conversion function
def molec_cm3(conc_kg_kg, mass, vol, mr):
    molec_box = conc_kg_kg*mass/mr*1000*n_a #molecules per box
    molec_cm3 = molec_box/(vol*1e6) #molec per cm3
    return molec_cm3
n_a = 6.02e23
mr_oh = 17
```

```python
# Import [OH] data
oh_949 = xr.open_dataarray('Data/bl949_oh_conc.nc',chunks={'time':24})
oh_443_data = xr.open_dataset(f'/gws/nopw/j04/gcp_ch4/u-bh443/u-bh443_301yrs_oh.nc',chunks={'time':24})
oh_443_kg = oh_443_data.oh

airm = xr.open_dataset(f'/gws/nopw/j04/gcp_ch4/u-bh443/u-bh443_301yrs_airmass_atm.nc',chunks={'time':24})
airm = airm.airmass_atm

alt = xr.open_dataset(f'/gws/nopw/j04/gcp_ch4/u-bl949/vol_n96.nc')
vol = alt.grid_cell_volume.data
```

```python
# Convert 443 data
oh_443 = molec_cm3(oh_443_kg,airm,vol,mr_oh)
```

```python
# Time average data, cannot use dask average as it removes the coordinates from the variables, preventing .loc from working in next step
oh_949_mean = np.mean(oh_949,axis=0)
oh_443_mean = np.mean(oh_443,axis=0)
```

```python
# Select longitudes wanted and remove altitudes above 20km
Alim = slice(0,55)

# Middle pacific           (no changes here, see what the agreement is)
lmPac = slice(180,190)
mPac_949 = dd.array.average(oh_949_mean.loc[Alim,:,lmPac],axis=2)
mPac_443 = dd.array.average(oh_443_mean.loc[Alim,:,lmPac],axis=2)
dmPac = mPac_949 - mPac_443

# Indian ocean (west)      (Changes here, look at distribution)
lwInd = slice(65,75)
wInd_949 = dd.array.average(oh_949_mean.loc[Alim,:,lwInd],axis=2)
wInd_443 = dd.array.average(oh_443_mean.loc[Alim,:,lwInd],axis=2)
dwInd = wInd_949 - wInd_443

# Middle Atlantic          (Changes through middle, try to determine nature of changes due to potentially unexpected large increase)
lmAtl = slice(315,335)
mAtl_949 = dd.array.average(oh_949_mean.loc[Alim,:,lmAtl],axis=2)
mAtl_443 = dd.array.average(oh_443_mean.loc[Alim,:,lmAtl],axis=2)
dmAtl = mAtl_949 - mAtl_443

# Middle Africa            (Relatively large changes, one up and one down, determine how high this goes too and compare to prescribed model)
lmAfr = slice(10,35)
mAfr_949 = dd.array.average(oh_949_mean.loc[Alim,:,lmAfr],axis=2)
mAfr_443 = dd.array.average(oh_443_mean.loc[Alim,:,lmAfr],axis=2)
dmAfr = mAfr_949 - mAfr_443
```

```python
# Plot on subplots in the same format as for [OH] maps, compare changes in different regions maybe using pacific as a starting point as an area unaffected by methane emissions

fig, axs = plt.subplots(nrows=4,ncols=3, figsize=(15,13),dpi=300)
fig.suptitle('Comparison of zonal mean [OH] in longitudes with the most change between 443 and 949',fontsize=14,y=0.95)

axs[0,0].set_title('u-bh443')
axs[0,0].set_ylabel('Middle Pacific') 
c1 = axs[0,0].pcolormesh(oh_443.latitude,alt.level_height[Alim],mPac_443,cmap='RdPu',shading='nearest',vmin=0,vmax=4e6)
cb1 = fig.colorbar(c1, label= '[OH] / molec. $cm^{-3}$', orientation="horizontal",extend='max',ax=axs[3,0])

axs[0,1].set_title('u-bl949')
c2 = axs[0,1].pcolormesh(oh_443.latitude,alt.level_height[Alim],mPac_949,cmap='RdPu',shading='nearest',vmin=0,vmax=4e6)
cb2 = fig.colorbar(c2, label= '[OH] / molec. $cm^{-3}$', orientation="horizontal",extend='max',ax=axs[3,1])

axs[0,2].set_title('Difference') 
c3 = axs[0,2].pcolormesh(oh_443.latitude,alt.level_height[Alim],dmPac,cmap='RdBu_r',shading='nearest',vmin=-6e4,vmax=6e4)
cb3 = fig.colorbar(c3, label= '[OH] / molec. $cm^{-3}$', orientation="horizontal",extend='both',ax=axs[3,2])


axs[1,0].set_ylabel('West Indian Ocean') 
axs[1,0].pcolormesh(oh_443.latitude,alt.level_height[Alim],wInd_443,cmap='RdPu',shading='nearest',vmin=0,vmax=4e6)

axs[1,1].pcolormesh(oh_443.latitude,alt.level_height[Alim],wInd_949,cmap='RdPu',shading='nearest',vmin=0,vmax=4e6)

axs[1,2].pcolormesh(oh_443.latitude,alt.level_height[Alim],dwInd,cmap='RdBu_r',shading='nearest',vmin=-6e4,vmax=6e4)


axs[2,0].set_ylabel('Mid Atlantic Ocean') 
axs[2,0].pcolormesh(oh_443.latitude,alt.level_height[Alim],mAtl_443,cmap='RdPu',shading='nearest',vmin=0,vmax=4e6)

axs[2,1].pcolormesh(oh_443.latitude,alt.level_height[Alim],mAtl_949,cmap='RdPu',shading='nearest',vmin=0,vmax=4e6)

axs[2,2].pcolormesh(oh_443.latitude,alt.level_height[Alim],dmAtl,cmap='RdBu_r',shading='nearest',vmin=-6e4,vmax=6e4)


axs[3,0].set_ylabel('Middle African Continent') 
axs[3,0].pcolormesh(oh_443.latitude,alt.level_height[Alim],mAfr_443,cmap='RdPu',shading='nearest',vmin=0,vmax=4e6)

axs[3,1].pcolormesh(oh_443.latitude,alt.level_height[Alim],mAfr_949,cmap='RdPu',shading='nearest',vmin=0,vmax=4e6)

axs[3,2].pcolormesh(oh_443.latitude,alt.level_height[Alim],dmAfr,cmap='RdBu_r',shading='nearest',vmin=-6e4,vmax=6e4)
```
